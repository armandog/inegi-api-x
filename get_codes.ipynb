```python
import pandas as pd
import requests

def get_inegi_meta(source, obj_name, obj_value, token):
  inegi_api = 'https://inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml'
  url_meta = f'{inegi_api}/{obj_name}/{obj_value}/es/{source}/2.0/{token}?type=json'

  response = requests.get(url_meta)
  
  if response.status_code == 200:
    data = response.json()
    series = data['CODE']
    df = pd.DataFrame(series)
  else:
    print('Error in request:', response.status_code)

  return(df)
```

```python
source = 'BIE-BISE'
token = 'd0975cf5-be76-1137-8b46-3e9a7ab9c5ed'

codes_df = get_inegi_meta(source, 'CL_INDICATOR', 'null', token)
codes_df.to_parquet('./codes/indicator_codes.parquet')

codes_df = get_inegi_meta(source, 'CL_FREQ', 'null', token)
codes_df.to_parquet('./codes/freq_codes.parquet')

codes_df = get_inegi_meta(source, 'CL_TOPIC', 'null', token)
codes_df.to_parquet('./codes/topic_codes.parquet')

codes_df = get_inegi_meta(source, 'CL_UNIT', 'null', token)
codes_df.to_parquet('./codes/unit_codes.parquet')

codes_df = get_inegi_meta(source, 'CL_STATUS', 'null', token)
codes_df.to_parquet('./codes/status_codes.parquet')
```
