```python
import requests

def get_inegi_data(source, indicator, geo_area, token, recent_data = True):
  inegi_api = 'https://inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml'

  if recent_data:
    recent = 'true'
  else:
    recent = 'false'

  url_data = f'{inegi_api}/INDICATOR/{indicator}/es/{geo_area}/{recent}/{source}/2.0/{token}?type=json'

  response = requests.get(url_data)
  
  if response.status_code == 200:
    data = response.json()
    series = data['Series'][0]['OBSERVATIONS']
    df = pd.DataFrame(series)
  else:
    print('Error in request:', response.status_code)

  return(data)

def get_inegi_meta(source, obj_name, obj_value, token):
  inegi_api = 'https://inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml'
  url_meta = f'{inegi_api}/{obj_name}/{obj_value}/es/{source}/2.0/{token}?type=json'

  response = requests.get(url_meta)
  
  if response.status_code == 200:
    data = response.json()
    series = data['CODE']
    df = pd.DataFrame(series)
  else:
    print('Error in request:', response.status_code)

  return(df)
```
